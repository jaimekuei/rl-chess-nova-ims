```python
import pickle
import os
from utils import load_checkpoint
import pandas as pd
import seaborn as sns
```

```python
strategy = 'dqn_white_stockfish_v2'
version = 'v4'

path_metrics = os.path.join(
    os.getcwd(), 'checkpoint', strategy, version, 
    'history_metrics', 'checkpoint_game_metrics_iteration_320.pkl')

with open(path_metrics, 'rb') as f:
    metrics = pickle.load(f)

results = pd.DataFrame(metrics)
results.plot.line(x='game', y='avg_plays', title='Q-Learning vs Stockfish - Average Plays')
sns.regplot(x='game', y='avg_plays', data=results)

results['window_avg_plays_sum'] = results['avg_plays'].rolling(10).mean()
results.plot.line(x='game', y='window_avg_plays_sum', title='Q-Learning vs Stockfish - Average Plays')
sns.regplot(x='game', y='window_avg_plays_sum', data=results)

results.plot.line(x='game', y='last_reward')
sns.regplot(x='game', y='last_reward', data=results)

results.plot.line(x='game', y='cummulated_reward')
sns.regplot(x='game', y='cummulated_reward', data=results)

results['window_cummulated_reward'] = results['cummulated_reward'].rolling(10).mean()
results.plot.line(x='game', y='window_cummulated_reward', title='Q-Learning vs Stockfish - Cummulated Reward')
sns.regplot(x='game', y='window_cummulated_reward', data=results)

results['process_time_w'] = results['process_time'].rolling(100).mean()
results.plot.line(x='game', y='process_time_w', title='Q-Learning vs Stockfish - Process Time')
sns.regplot(x='game', y='process_time_w', data=results)
```
